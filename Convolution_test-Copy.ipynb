```python
import skimage.data
import matplotlib.pyplot as plt
import numpy as np
# Reading the image
kernel=np.array([[-1,-1,-1],[-1,4,-1],[-1,-1,-1]])
img = skimage.data.chelsea()
# Converting the image into gray.
img = skimage.color.rgb2gray(img)
img1=np.zeros((302,453))
img1[1:301,1:452]=img
img2=np.zeros((302,453))
for i in range(1,301):
    for j in range(1,452):
        img2[i,j]=sum(sum(img1[i-1:i+2,j-1:j+2]*kernel))/9
plt.imshow(img2,cmap='gray')
```

```python
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import normalize
data=pd.read_csv('headbrain.csv')
X=data['col1'].values
Y=data['col2'].values
Z=data['col3'].values
Z=Z.reshape(-1,1)
Z=normalize(Z)
a=pd.DataFrame(columns=['col1','col2','col3'])
for i in range(0,X.shape[0]):
    a.loc[i]=[X[i],Y[i],Z[i]]
y=data['col4'].values
y=y.reshape(-1,1)
y=normalize(y)
X_train,X_test,y_train,y_test=train_test_split(a,y,test_size=0.1)
model=Sequential()

model.add(Dense(4,input_dim=3,activation='relu'))
model.add(Dense(3,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
model.compile(metrics=['accuracy'],optimizer='adam',loss='binary_crossentropy')  
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=4)
model.predict(a[:4])
```
